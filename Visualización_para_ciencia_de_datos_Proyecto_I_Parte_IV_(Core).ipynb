```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv('/content/retail_sales_dataset.csv')
```

### Análisis Descriptivo y Visualización
* Calcula estadísticas descriptivas básicas para comprender mejor las características de los datos.
* Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.
* Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.
* Utiliza gráficos de dispersión para analizar la relación entre diferentes variables.
* Combina histogramas y boxplots para visualizar la distribución de las ventas y detectar valores atípicos.

```python
# Convertir la columna de fecha al tipo de datos datetime
data['Date'] = pd.to_datetime(data['Date'])

# 1. Estadísticas descriptivas básicas
print("Estadísticas descriptivas:")
print(data.describe())
```

```python
# 2. Visualización de la distribución de variables numéricas clave

# Histograma de la edad
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Boxplot para la cantidad de productos vendidos
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Quantity'], color='lightgreen')
plt.title('Distribución de la Cantidad de Productos Vendidos')
plt.xlabel('Cantidad')
plt.show()
```

```python
# 3. Gráfico de líneas para mostrar tendencias de ventas a lo largo del tiempo

# Crear una columna 'Mes' para agrupar por mes
data['Month'] = data['Date'].dt.to_period('M')
monthly_sales = data.groupby('Month')['Total Amount'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='b')
plt.title('Tendencias de Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Monto Total de Ventas')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Gráficos de dispersión para analizar relaciones entre variables

# Relación entre Edad y Monto Total de la compra
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Total Amount', data=data, color='purple')
plt.title('Relación entre Edad y Monto Total de la Compra')
plt.xlabel('Edad')
plt.ylabel('Monto Total')
plt.show()

# Relación entre Cantidad y Monto Total
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Total Amount', data=data, color='orange')
plt.title('Relación entre Cantidad y Monto Total')
plt.xlabel('Cantidad')
plt.ylabel('Monto Total')
plt.show()
```

```python
# 5. Combinación de histogramas y boxplots para analizar la distribución y detectar valores atípicos

# Distribución de Monto Total con histograma y boxplot
plt.figure(figsize=(14, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(data['Total Amount'], bins=15, color='lightcoral', edgecolor='black')
plt.title('Distribución del Monto Total de Ventas')
plt.xlabel('Monto Total')
plt.ylabel('Frecuencia')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=data['Total Amount'], color='lightblue')
plt.title('Boxplot del Monto Total de Ventas')
plt.xlabel('Monto Total')

plt.show()
```

#Explicación del Código
* **Estadísticas descriptivas**: Calculamos estadísticas básicas como media, desviación estándar, etc., para tener una idea inicial de los datos.
* **Distribuciones:** Utilizamos histogramas para observar la distribución de la edad y boxplots para ver la cantidad de productos vendidos, lo que ayuda a identificar valores atípicos.
* **Tendencias de ventas:** Usamos un gráfico de líneas de ventas totales agrupadas por mes para identificar patrones de crecimiento o estacionalidad.
* **Relación entre variables:** Los gráficos de dispersión nos ayudan a ver la relación entre edad y monto total, así como entre cantidad y monto total, lo que puede sugerir patrones entre estas variables.
* **Detección de valores atípicos:** Combinamos histogramas y boxplots para analizar la distribución del monto total y detectar posibles valores atípicos visualmente.
